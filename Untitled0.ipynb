```python
# Import Libraries
import pandas as pd
import lasio
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Print LAS curves
las =lasio.read("/content/177074051000_3620.las")
```

```python
print(las.curves)
```

```python
df=las.df()
print(df)
```

```python
df.head()
```

```python
#Plot GR log
plt.figure(figsize=(4,8))
plt.plot(df["DT"], df.index)
plt.gca().invert_yaxis()
plt.xlabel("DT")
plt.ylabel("Depth")
plt.title("DT Log")
plt.show()
```
